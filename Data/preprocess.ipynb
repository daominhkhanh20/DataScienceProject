# Transform date
    * Only select month

```python
dataset
```

```python
def transform_data(dataset):
    datetime = dataset["createdAt"]
    year = []
    month = []
    for x in datetime:
        if x[-2:] == "·ªõc":
            if "th√°ng" in x:
                time_duration = int(x.split(" ")[0])
                create_month = 11 - time_duration
                if create_month < 1:
                    create_month += 12
                    create_year = 2020
                else:
                    create_year = 2021
            else:
                create_month = 11
                create_year = 2021
            year.append(create_year)
            month.append(create_month)
        elif x == "h√¥m qua":
            month.append(11)
            year.append(2021)
        else:
            print(x.split('/'))
            month_, year_ = x.split("/")[1:]
            month.append(int(month_))
            year.append(int(year_))
    return month, year
```

```python
month, year = transform_data(dataset)
```

## Note: year only contain "2021"

```python
unique_list = []

for x in year:
    if x in unique_list:
        continue
    else:
        unique_list.append(x)
```

```python
dataset['month'] = month
```

```python
dataset = dataset.drop(["createdAt"], axis = 1)
```
