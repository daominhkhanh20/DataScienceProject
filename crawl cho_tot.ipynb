```python
import requests
import pandas as pd
```

```python
#o :post item timestamp
```

```python
response = requests.get("https://gateway.chotot.com/v1/public/ad-listing?cg=2010&o=0")
```

```python
maxIndex = 29773
```

```python
cols = []
for x in response.json()['ads'][0]:
    cols.append(x)
data= pd.DataFrame(columns=cols)
for x in range(1000): #100 times
    if(x%10==0):
        print(x)
    url = "https://gateway.chotot.com/v1/public/ad-listing?cg=2010&o=" + str(x)
    response = requests.get(url)
    for i in range(len(response.json()['ads'])):
        ads= response.json()['ads'][i]
        data = data.append(ads,ignore_index=True)
```
