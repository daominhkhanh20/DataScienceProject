```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_json('../Data/hang.json')
```

```python
data
```

```python
res = pd.DataFrame(data.data[0])
```

```python
res
```

```python
#filter only web bonbanh
```

```python
res = res[res['url'].str[8:15] =='bonbanh']
```

```python
res.reset_index(inplace=True,drop=True)
```

```python
res
```

```python
def dict_to_frame(data,colName):
    data = list(data[colName])
    return pd.DataFrame(data)
```

```python
res1 = dict_to_frame(res,'category')
res1
```

```python
res2 = dict_to_frame(res1,'source')
res2
```

```python
res3 = dict_to_frame(res1,'car_model')
res3
```

```python
result = pd.concat([res,res1[['id','car_model_id','source_id']],res3],axis=1)
```

```python
result.drop(columns='category',inplace=True)
```

```python
result.to_csv('../Data/data_bonbanh.csv',index=False)
```
